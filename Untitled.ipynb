```python
%matplotlib notebook
import mne
import numpy as np
import matplotlib.pyplot as plt
sfreq=500
chan=1
```

```python
info = mne.create_info(['EEG 00' + str(idx+1) for idx in range(1000)],
                       sfreq, ch_types='eeg')
```

```python
ul=80 # smallest output mV
uh=220 # highest output mV
nl=1
nu=1
zu=[]
cl=np.zeros((1000,1))
cu=np.zeros((1000,1))
amplitude=np.zeros((1000,1))
data = np.random.randn(1000,500)+5
for kk in range(1000):
    raw = mne.io.RawArray(data, info);
    psds, freqs = mne.time_frequency.psd_welch(raw.pick_channels([info['ch_names'][kk]]), fmin=4, fmax=45, n_fft=256);
    freq_idxs = np.where((freqs > 8) & (freqs < 13))[0];
    s=np.mean(np.sqrt(psds[0][freq_idxs]));
    if kk < 100:
        zu.append(s)
    else:
        nu=max(s,nu);
        nl=min(s,nl);
        amplitude[kk] = (s-nl)/(nu-nl)*(uh/ul-1)+1;
        cu[kk]=nu;
        cl[kk]=nl;
print("Hello world!")
```

```python
fig = plt.figure()

#ax = fig.add_subplot(111)
#plt.plot(range(1000),amplitude*7.5)
plt.plot(range(100),zu)
print(min(zu))
print(max(zu))
```
